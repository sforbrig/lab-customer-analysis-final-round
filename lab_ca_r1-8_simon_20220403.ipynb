```python
#change chosen column names
file1 = file1.rename(columns={'ST':'State',
                            'GENDER':'Gender'})

#change columns to lower letters
cols = []
for column in file1.columns:
    cols.append(column.lower())
file1.columns = cols

#kick out space between words in columns and replace it with an underscore
file1.columns = file1.columns.str.replace(' ','_')

#kick out '%' in customer_lifetime_value, change string to numeric and divide with 100 (because of percentage before)
file1['customer_lifetime_value'] = file1['customer_lifetime_value'].replace('%', '', regex=True)
file1['customer_lifetime_value'] = pd.to_numeric(file1['customer_lifetime_value'], errors='coerce')
file1['customer_lifetime_value'] = file1['customer_lifetime_value'] /100

#check result by head
file1.head()
```

```python
#LAB2:
#Show the dataframe shape
mca = pd.read_csv('marketing_customer_analysis.csv')

display(mca.head())
display(mca.tail())

mca.shape
```

```python
#Standardize header names:

#change columns to lower letters
cols = []
for column in mca.columns:
    cols.append(column.lower())
mca.columns = cols

#kick out space between words in columns and replace it with an underscore
mca.columns = mca.columns.str.replace(' ','_')

#check result by head
file1.head()
```

```python
#Check and deal with NaN values.
mca.isna().sum()
```

```python
round(mca.isna().sum()/len(mca),4)*100
nulls_mca = pd.DataFrame(round(mca.isna().sum()/len(mca),4)*100)
nulls_mca
nulls_mca = nulls_mca.reset_index()
nulls_mca
nulls_mca.columns = ['header_name', 'percent_nulls']
nulls_mca
```

```python
#check unique values in column vehicle_type
mca['vehicle_type'].unique()
```

```python
#change NaN to B becaues only A as value avalible
mca['vehicle_type'] = mca['vehicle_type'].fillna('B')

mca['vehicle_type'].value_counts(dropna=False)
```

```python
#kick out unnamed:_0
mca=mca.drop(columns=['unnamed:_0'])
```

```python
import statistics as stats
```

```python
#use mode to replace the NaN for values in other columns 
mca['state'] = mca['state'].fillna(stats.mode(mca['state']))
mca['response'] = mca['response'].fillna(stats.mode(mca['response']))
mca['months_since_last_claim'] = mca['months_since_last_claim'].fillna(stats.mode(mca['months_since_last_claim']))
mca['number_of_open_complaints'] = mca['number_of_open_complaints'].fillna(stats.mode(mca['number_of_open_complaints']))
mca['vehicle_class'] = mca['vehicle_class'].fillna(stats.mode(mca['vehicle_class']))
mca['vehicle_size'] = mca['vehicle_size'].fillna(stats.mode(mca['vehicle_size']))
```

```python
#check again NaN:
mca.isna().sum()
```

```python
#change 'effective_to_date' to type datetime 
mca['effective_to_date_new'] = pd.to_datetime(mca['effective_to_date'], errors='coerce')

mca.dtypes
```

```python
#change 'effective_to_date_new' to show just month
mca['effective_to_date_new'] = pd.DatetimeIndex(mca['effective_to_date_new']).month
mca.head()
```

```python
#filter the data to show only the information for the first quarter
mca_QI = mca[mca['effective_to_date_new']<4]
display(mca_QI.head(20))
display(mca_QI.tail(20))
```

```python
#bin into categories by function:
def tca_bins(x):
    if x <=500:
        return 1
    elif x<=1000:
        return 2
    elif x<=1500:
        return 3
    elif x<=2000:
        return 4
    else:
        return 5
    
mca_QI['total_claim_amount_bins'] = list(map(tca_bins, mca_QI['total_claim_amount']))

mca_QI
```

```python
def income_bins(x):
    if x <20000:
        return 1
    elif x<40000:
        return 2
    elif x<60000:
        return 3
    elif x<80000:
        return 4
    else:
        return 5
    
mca_QI['income_bins'] = list(map(income_bins, mca_QI['income']))

mca_QI
```
